```
df = pd.read_csv('google_review_ratings.csv') 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import scipy.stats
del(df['User'])
del(df['Unnamed: 25'])
df=df.replace('2\t2.',2.2)
df=df.replace(np.NaN,0)
df['Category 11']=df['Category 11'].astype(float)
print(df['Category 11'])
print(df['Category 1'])
```

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import scatter_matrix

kmeans=KMeans(n_clusters=5,random_state=1012)


result=kmeans.fit_predict(df)
df['cluster_id']=result
df['cluster_id'].value_counts()
```

```
import matplotlib.pyplot as plt
 
clusterinfo = pd.DataFrame()
for i in range(5):
    clusterinfo['cluster' + str(i)] = df[df['cluster_id'] == i].mean()
clusterinfo = clusterinfo.drop('cluster_id')
my_plot1 = clusterinfo.plot(kind='bar', stacked=True)
my_plot = clusterinfo.T.plot(kind='bar', stacked=True, title="Mean Value of 5 Clusters")
```

```
#在cluster0 bar高:
c0=df[['Category 2','Category 5', 'Category 6','Category 7']]

#在cluster1 bar高:
c1=df[['Category 1', 'Category 16','Category 17', 'Category 18', 'Category 19', 'Category 20', 'Category 21', 'Category 23', 'Category 24']]

#在cluster2 bar高:
c2=df[['Category 12', 'Category 13', 'Category 14', 'Category 15']]

#在cluster3 bar高:
c3=df[['Category 8', 'Category 9', 'Category 10', 'Category 11']]

#在cluster4 bar高:
c4=df[['Category 3','Category 4','Category 22']]

```

```
#把cluster1 再細分
kmeans=KMeans(n_clusters=3,random_state=1012)


result=kmeans.fit_predict(c1)
c1['cluster_id']=result
c1['cluster_id'].value_counts()

clusterinfo1 = pd.DataFrame()
for i in range(3):
    clusterinfo1['cluster' + str(i)] = c1[c1['cluster_id'] == i].mean()
clusterinfo1 = clusterinfo1.drop('cluster_id')
my_plot1 = clusterinfo1.plot(kind='bar', stacked=True)
my_plot = clusterinfo1.T.plot(kind='bar', stacked=True)
```
